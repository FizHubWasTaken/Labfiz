```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from math import pi, tan
```

```
mi = 4 * pi * 1e-7
R = 0.13
uR = 0.0015
klasa = 0.005
```

```
def rad(deg):
  return deg * pi / 180
```

```
def B(tab):
  lst = []
  for i in range(len(tab['Liczba zwojów'])):
    N, I, a = tab['Liczba zwojów'][i], tab['Prąd I [A]'][i], tab['Średni kąt nachylenia [rad]'][i]
    lst.append(mi * N * I / (2 * R * tan(a)))
  return lst
```
